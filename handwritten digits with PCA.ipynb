### Importing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
digits = pd.read_csv('../DATA/digits.csv')
```

```python
digits
```

**Creating a new DataFrame called *pixels* that consists only of the pixel feature values by dropping the number_label column.**

```python
pixels = digits.drop('number_label',axis=1)
```

```python
pixels
```

### Scaling Data

**Using Scikit-Learn to scale the pixel feature dataframe.**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_pixels = scaler.fit_transform(pixels)

scaled_pixels
```

### PCA


```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

**variance explained by 2 principal components.**

```python
np.sum(pca_model.explained_variance_ratio_)
```

**Creating a scatterplot of the digits in the 2 dimensional PCA space, color/label based on the original number_label column in the original dataset.**

```python
plt.figure(figsize=(10,6),dpi=150)
labels = digits['number_label'].values
sns.scatterplot(pca_pixels[:,0],pca_pixels[:,1],hue=labels,palette='Set1')
plt.legend(loc=(1.05,0))
```

**Creating an "interactive" 3D plot of the result of PCA with 3 principal components.**

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=3)
```

```python
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

```python
from mpl_toolkits import mplot3d
```

```python
plt.figure(figsize=(8,8),dpi=150)
ax = plt.axes(projection='3d')
ax.scatter3D(pca_pixels[:,0],pca_pixels[:,1],pca_pixels[:,2],c=digits['number_label']);
```
